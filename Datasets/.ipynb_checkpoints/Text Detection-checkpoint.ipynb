```python
url = 'https://connectingsigns.com/wp-content/uploads/2019/06/one-way-road-sign.jpg'
```

```python
img = Image.open(urllib.request.urlopen(url))
img
```

```python
image = np.asarray(bytearray(urllib.request.urlopen(url).read()), dtype="uint8")
```

```python
image = cv2.imdecode(image, cv2.IMREAD_COLOR)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
filename = "{}.png".format(os.getpid())
cv2.imwrite(filename, gray)
```

```python
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
```

```python
ret,img = cv2.threshold(np.array(img), 125, 255, cv2.THRESH_BINARY)
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.findContours(np.array(img), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
img = Image.fromarray(img.astype(np.uint8))
```
